1. 이미지 읽기 옵션
   - cv2.IMREAD_COLOR : 컬러 이미지, 투명 영역은 무시(디폴트)
   - cv2.IMREAD_GRAYSCALE : 흑백 이미지
   - cv2.IMREAD_UNCHANGED : 투명 영역까지 포함

```python
import cv2
img_color = cv2.imread('img.jpg', cv2.IMREAD_COLOR)
img_gray = cv2.imread('img.jpg', cv2.IMREAD_GRAYSCALE)
img_unchaged = cv2.imread('img.jpg', cv2.IMREAD_UNCHANGED)

cv2.imshow('img_color', img_color)
cv2.imshow('img_gray', img_gray)
cv2.imshow('img_unchaged', img_unchaged)

if cv2.waitKey(0) == ord('A'):
    print('사용자가 A를 눌렀음.')
cv2.destroyAllWindows()
```

2. 동영상 보여주기
   - 동영상 파일 보여주기

```python
import cv2
cap = cv2.VideoCapture('video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print('더 이상 가져올 이미지가 없다.')
        break
    cv2.imshow('video' , frame)

    if cv2.waitKey(50) == ord('q'):
        print('사용자에 의해 종료됨.')
        break

cap.release()
cv2.destroyAllWindows()
```

3. 도형 그리기
   - 빈 스케치북 만들기

```python
import cv2
import numpy as np

# 세로 480 * 가로 640, 3 채널 [RGB] 에 해당하는 빈 스케치북 만들기
img = np.zeros((480,640,3), dtype=np.uint8)
# img[:] = [0,255,255]
# print(img)
img[200:300, 300:400] = (255,255,255)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 - 직선 그리기
   - cv2.LINE_4 : 상하좌우 4가지 방향으로만 연결된 선
   - cv2.LINE_8 : 대각선을 포함한 8가지 방향으로 연결된 선(디폴트)
   - cv2.LINE_AA : 부드러운 선(anti-aliasing)

- cv2.line( 그릴위치 , 시작점, 끝점, 색깔, 두께, 선 종류 )

```python
import cv2
import numpy as np

img = np.zeros((480,640,3), dtype=np.uint8)

COLOR = (0, 0, 255)
THIC = 3

cv2.line( img, (50, 50), (400, 400), COLOR, THIC , cv2.LINE_AA)
cv2.line( img, (400, 50), (50, 400), COLOR, THIC , cv2.LINE_AA)
# cv2.line( img, (50, 300), (400, 250), COLOR, THIC , cv2.LINE_AA)

cv2.imshow('line', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 - 원 그리기
   - cv2.circle( 그릴위치, 원의 중심점, 반지름, 색깔, 두께, 선종류)

```python
import cv2
import numpy as np

img = np.zeros((480,640,3), dtype=np.uint8)

COLOR = (255, 255, 0)
THIC = 10
RADIUS = 50

cv2.circle( img, (200, 100), RADIUS, COLOR, THIC , cv2.LINE_AA)
cv2.circle( img, (400, 100), RADIUS, COLOR, cv2.FILLED , cv2.LINE_AA)

cv2.imshow('circle', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

 - 사각형 그리기
   - cv2.rectangle( 그릴위치, 왼쪽위 좌표, 오른쪽 아래 좌표, 색깔, 두께 )

```python
import cv2
import numpy as np

img = np.zeros((480,640,3), dtype=np.uint8)

COLOR = (0, 255, 0)
THIC = 3

cv2.rectangle( img, (100, 100), (400, 400), COLOR, THIC)
cv2.rectangle( img, (200, 200), (300, 300), COLOR, cv2.FILLED)

cv2.imshow('rectangle', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

  - 다각형 그리기
    - cv2.polylines( 그릴위치, 그릴좌표들 , 닫힘여부(T/F) , 색깔, 두께, 선종류) : 비어있는 다각형
    - cv2.fillPoly( 그릴위치, 그릴좌표들 , 색깔, 선 종류)

```python
import cv2
import numpy as np

img = np.zeros((480,640,3), dtype=np.uint8)

COLOR = (0, 255, 0)
THIC = 3

pt1 = np.array([ [ [100, 100], [200, 100], [100, 200] ] , [ [200, 100], [300, 100], [300, 200] ] ])
# pt2 = np.array([ [200, 100], [300, 100], [300, 200] ])
cv2.polylines( img, pt1 , True , COLOR , THIC, cv2.LINE_AA)

pt2 = np.array([ [ [100, 300], [200, 300], [100, 400] ] , [ [200, 300], [300, 300], [300, 400] ] ])
cv2.fillPoly( img, pt2 ,  COLOR , cv2.LINE_AA)
# cv2.polylines( img, [pt2] , True , COLOR , THIC, cv2.LINE_AA)


cv2.imshow('polylines', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

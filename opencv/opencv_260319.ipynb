```python
import cv2
import numpy as np

# 1. 초기 설정
width, height = 800, 600 # x가 움직일 공간을 위해 가로를 조금 넓혔어요
color = (0, 200, 0)
radius = 30
rad_direction = 1

# x좌표 이동 관련 변수
offset_x = 0
x_direction = 1 # 1이면 오른쪽, -1이면 왼쪽

# 2. 원본 좌표 데이터 (0 위치 기준)
fill_list_raw = [
    np.array([[[100,100],[200,100],[100,200]], [[200,100],[300,100],[300,200]]]),
    np.array([[[400,100],[500,100],[500,200]], [[500,200],[500,300],[400,300]]]),
    np.array([[[300,400],[300,500],[400,500]], [[400,500],[500,500],[500,400]]]),
    np.array([[[100,300],[200,300],[100,400]], [[100,400],[100,500],[200,500]]])
]

line_list_raw = [
    np.array([[[100,200],[100,300],[200,300]], [[200,300],[300,300],[300,200]]]),
    np.array([[[300,100],[400,100],[300,200]], [[300,200],[300,300],[400,300]]]),
    np.array([[[300,300],[300,400],[400,300]], [[400,300],[500,300],[500,400]]]),
    np.array([[[200,300],[300,300],[300,400]], [[300,400],[300,500],[200,500]]])
]

circle_centers_raw = [(200, 200), (400, 200), (200, 400), (400, 400)]
circle_colors = [(200, 200, 0), (200, 100, 100), (100, 200, 100), (0, 200, 200)]

while True:
    img = np.zeros((height, width, 3), dtype=np.uint8)

    # 3. 모든 도형에 offset_x 적용 (Numpy 브로드캐스팅 활용)
    # 각 폴리곤 좌표에 [offset_x, 0]을 더해서 x축으로만 이동시킴
    for poly in fill_list_raw:
        moved_poly = poly + [offset_x, 0]
        cv2.fillPoly(img, moved_poly, color, cv2.LINE_AA)

    for poly in line_list_raw:
        moved_poly = poly + [offset_x, 0]
        cv2.polylines(img, moved_poly, True, color, 1, cv2.LINE_AA)

    # 4. 원 좌표 업데이트 및 그리기
    for i in range(4):
        cx, cy = circle_centers_raw[i]
        cv2.circle(img, (cx + offset_x, cy), radius, circle_colors[i], cv2.FILLED, cv2.LINE_AA)

    # 5. 상태 업데이트
    # 반지름 애니메이션
    radius += rad_direction
    if radius >= 50 or radius <= 30: rad_direction *= -1

    # X축 이동 애니메이션 (0 ~ 50)
    offset_x += x_direction
    if offset_x >= 200 or offset_x <= 0:
        x_direction *= -1

    cv2.imshow('Moving Shapes', img)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(1) # macOS 전용 버그 픽스용 한 번 더 호출
```

```python
import cv2
import numpy as np

# 1. 초기 설정 (이전과 동일)
width, height = 800, 600 
color = (0, 200, 0)
radius = 30
rad_direction = 1

# x좌표 이동 관련 변수 (0 ~ 100으로 범위를 좀 넓혔어요)
offset_x = 0
x_direction = 1 # 1이면 오른쪽, -1이면 왼쪽

# 2. 원본 좌표 데이터 (이전 코드의 데이터를 그대로 사용하세요)
# (pt1 ~ pt8 좌표 데이터는 생략할게요. 위 코드의 raw 데이터를 그대로 넣어주세요.)
fill_list_raw = [
    np.array([[[100,100],[200,100],[100,200]], [[200,100],[300,100],[300,200]]]), # pt1
    np.array([[[400,100],[500,100],[500,200]], [[500,200],[500,300],[400,300]]]), # pt4
    np.array([[[300,400],[300,500],[400,500]], [[400,500],[500,500],[500,400]]]), # pt6
    np.array([[[100,300],[200,300],[100,400]], [[100,400],[100,500],[200,500]]])  # pt7
]

line_list_raw = [
    np.array([[[100,200],[100,300],[200,300]], [[200,300],[300,300],[300,200]]]), # pt2
    np.array([[[300,100],[400,100],[300,200]], [[300,200],[300,300],[400,300]]]), # pt3
    np.array([[[300,300],[300,400],[400,300]], [[400,300],[500,300],[500,400]]]), # pt5
    np.array([[[200,300],[300,300],[300,400]], [[300,400],[300,500],[200,500]]])  # pt8
]

circle_centers_raw = [(200, 200), (400, 200), (200, 400), (400, 400)]
circle_colors = [(200, 200, 0), (200, 100, 100), (100, 200, 100), (0, 200, 200)]

while True:
    img = np.zeros((height, width, 3), dtype=np.uint8)

    # 3. 중요!! 슬라이싱을 활용한 지그재그 그리기
    
    # --- 채울 도형들 (Fill) ---
    # 홀수 번째 (index 0, 2): 오른쪽으로 이동 (+ offset_x)
    for poly in fill_list_raw[::2]:
        moved_poly = poly + [offset_x, 0]
        cv2.fillPoly(img, moved_poly, color, cv2.LINE_AA)
        
    # 짝수 번째 (index 1, 3): 왼쪽으로 이동 (- offset_x)
    for poly in fill_list_raw[1::2]:
        moved_poly = poly - [offset_x, 0]
        cv2.fillPoly(img, moved_poly, color, cv2.LINE_AA)

    # --- 테두리 도형들 (Line) ---
    # 홀수 번째: 오른쪽으로 이동
    for poly in line_list_raw[::2]:
        moved_poly = poly + [offset_x, 0]
        cv2.polylines(img, moved_poly, True, color, 1, cv2.LINE_AA)
        
    # 짝수 번째: 왼쪽으로 이동
    for poly in line_list_raw[1::2]:
        moved_poly = poly - [offset_x, 0]
        cv2.polylines(img, moved_poly, True, color, 1, cv2.LINE_AA)

    # --- 원 그리기 ---
    # 홀수 번째 원 (0, 2): 오른쪽
    for i in [0, 2]:
        cx, cy = circle_centers_raw[i]
        cv2.circle(img, (cx + offset_x, cy), radius, circle_colors[i], cv2.FILLED, cv2.LINE_AA)
        
    # 짝수 번째 원 (1, 3): 왼쪽
    for i in [1, 3]:
        cx, cy = circle_centers_raw[i]
        cv2.circle(img, (cx - offset_x, cy), radius, circle_colors[i], cv2.FILLED, cv2.LINE_AA)

    # 4. 상태 업데이트
    radius += rad_direction
    if radius >= 50 or radius <= 30: rad_direction *= -1

    # X축 이동 애니메이션 (0 ~ 100)
    offset_x += x_direction
    if offset_x >= 100 or offset_x <= 0:
        x_direction *= -1

    cv2.imshow('ZigZag Shapes', img)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cv2.waitKey(1) # macOS 전용 버그 픽스용 한 번 더 호출
```

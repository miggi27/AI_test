### 텍스트
- OpenCV에서 사용하는 글꼴 종류
  - cv2.FONT_HERSHEY_SIMPLEX : 보통 크기의 산 세리프(sans-serif)
  - cv2.FONT_HERSHEY_PLAIN : 작은 크기의 산 세리프 글꼴
  - cv2.FONT_HERSHEY_SCRIPT_SIMPLEX : 필기체 스타일 글꼴
  - cv2.FONT_HERSHEY_TRIPLEX : 보통 크기의 세리프 글꼴
  - cv2.FONT_ITALIC : 기울임

### cv2.putText(그릴위치, 텍스트내용, 시작위치, 폰트종류, 크기, 색깔, 두께)

```python
import cv2
import numpy as np

img = np.zeros((480, 640, 3), dtype=np.uint8)

scale = 1
color = (255, 255, 255)
thic = 1

cv2.putText(img, "Korea Fighting!!", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, scale, color, thic)
cv2.putText(img, "Korea Fighting!!", (20, 150), cv2.FONT_HERSHEY_PLAIN, scale, color, thic)
cv2.putText(img, "Korea Fighting!!", (20, 250), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, scale, color, thic)
cv2.putText(img, "Korea Fighting!!", (20, 350), cv2.FONT_HERSHEY_TRIPLEX, scale, color, thic)

# 핵심 수정: 기존 폰트와 이탤릭 플래그를 조합해야 합니다.
cv2.putText(img, "Korea Fighting!!", (20, 450), cv2.FONT_HERSHEY_SIMPLEX | cv2.FONT_ITALIC, scale, color, thic)

cv2.imshow('text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import numpy as np
# PIL(python image library)
from PIL import ImageFont, ImageDraw, Image

def myPutText(src, text, pos, font_size, font_color):
    img_pil = Image.fromarray(src)
    draw = ImageDraw.Draw(img_pil)
    font = ImageFont.truetype('fonts/gulim.ttc', font_size)
    draw.text(pos, text, font=font, fill=font_color)
    return np.array(img_pil)

img = np.zeros((480, 640, 3), dtype=np.uint8)

font_size = 30
color = (255, 255, 255)
# thic = 1

# cv2.putText(img, "Korea Fighting!!", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, scale, color, thic)
img = myPutText(img, "대한민국 화이팅!!", (20, 50), font_size, color)

cv2.imshow('text', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from PIL import ImageFont, ImageDraw, Image

# 1. 빈 이미지 생성
img = np.zeros((480, 640, 3), dtype=np.uint8)

# 2. 사용할 폰트 경로 설정 (윈도우 기본 폰트 경로)
# 이탤릭체를 위해 'malgunsl.ttf' (맑은 고딕 Semilight/Italic 등)를 쓰거나 
# 일반 폰트를 불러와서 처리합니다.
font_path = "C:/Windows/Fonts/malgun.ttf" # 일반 맑은 고딕
font_italic_path = "C:/Windows/Fonts/verdanai.ttf" # 베르다나 이탤릭 (영문용)

# 3. OpenCV 이미지를 PIL 이미지로 변환
img_pil = Image.fromarray(img)
draw = ImageDraw.Draw(img_pil)

# 4. 폰트 설정 (폰트 파일, 크기)
font_kor = ImageFont.truetype(font_path, 40)
font_eng_italic = ImageFont.truetype(font_italic_path, 40)

# 5. 글자 쓰기 (좌표, 내용, 폰트, 색상)
draw.text((50, 50), "대한민국 파이팅!! (한글 성공)", font=font_kor, fill=(255, 255, 255))
draw.text((50, 150), "Korea Fighting!! (Italic)", font=font_eng_italic, fill=(255, 255, 0))

# 6. 다시 OpenCV 이미지(numpy array)로 변환
img = np.array(img_pil)

# 7. 출력
cv2.imshow('Hangul and Italic', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
cap = cv2.VideoCapture('cat.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow('video', frame)
    if cv2.waitKey(20) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture('cat.mp4')

## 코덱을 정의하는 부분
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter('output.avi', fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    out.write(frame)
    cv2.imshow('video', frame)
    if cv2.waitKey(20) == ord('q'):
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```

### 이미지 크기 조정
- 이미지
  - 고정 크기로 설정

```python
import cv2
img = cv2.imread('cat.jpg')
# 고정 크기 변경
dst = cv2.resize(img, (400, 500))

cv2.imshow('img', img)
cv2.imshow('resize', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('cat.jpg')
# 비율 크기 변경
dst = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
# dst = cv2.resize(img, (400, 500))

cv2.imshow('img', img)
cv2.imshow('resize', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 보간법
1. cv2.INTER_AREA : 크기를 줄일 때
2. cv2.INTER_CUBIC : 크기를 늘릴 때 사용 (퀄리티 좋음, 속도 느림)
3. cv2.INTER_LINEAR : 크기를 늘릴 때 사용(디폴트)

- 동영상
  - 고정크기로 설정

```python
import cv2
cap = cv2.VideoCapture('cat.mp4')
# 고정 크기 변경
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    frame_resize = cv2.resize(frame, (400, 500))
    cv2.imshow('resize', frame_resize)
    if cv2.waitKey(20) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

  - 비율로 설정

```python
import cv2
cap = cv2.VideoCapture('cat.mp4')
# 비율 크기 변경
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_resize = cv2.resize(frame, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
    cv2.imshow('resize', frame_resize)
    if cv2.waitKey(20) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

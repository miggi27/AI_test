### 이미지 자르기

- 이미지 잘라서 새로운 창에 표시

```python
import cv2
img = cv2.imread('img.jpg')
# img.shape  (427, 640, 3)
crop = img[81:351, 31:285]   # 세로 기준값 , 가로 기준값

cv2.imshow('img', img)
cv2.imshow('crop', crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 자른 이미지를 원본이미지에 합성

```python
img = cv2.imread('img.jpg')
# img.shape  (427, 640, 3)
crop = img[81:351, 31:285]   # 세로 기준값 , 가로 기준값
img[81:351, 335:589] = crop

cv2.imshow('img', img)
# cv2.imshow('crop', crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

이미지 대칭

- 좌우 대칭

```python
img = cv2.imread('img.jpg')

flip_hori = cv2.flip(img, 1)  # flipcode > 0 : 수평

cv2.imshow('img', img)
cv2.imshow('flip_hori', flip_hori)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 상하대칭

```python
img = cv2.imread('img.jpg')

flip_ver = cv2.flip(img, 0)  # flipcode == 0 : 수직

cv2.imshow('img', img)
cv2.imshow('flip_ver', flip_ver)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 상하좌우 대칭

```python
img = cv2.imread('img.jpg')

flip_both = cv2.flip(img, -1)  # flipcode < 0 : 수직수평

cv2.imshow('img', img)
cv2.imshow('flip_both', flip_both)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 회전

```python
img = cv2.imread('img.jpg')

rotate90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) # 시계방향 90도 회전
rotate180 = cv2.rotate(img, cv2.ROTATE_180)  # 180도 회전
rotate270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) # 시계반대방향 90도

cv2.imshow('img', img)
cv2.imshow('rotate90', rotate90)
cv2.imshow('rotate180', rotate180)
cv2.imshow('rotate270', rotate270)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 블러(흐리게)

- 커널 사이즈 변화에 따른 블러 ( 가우시안 블러 )

```python
img = cv2.imread('img.jpg')

# 커널 사이즈 => (3, 3) (5, 5) (7, 7)
kenel_33 = cv2.GaussianBlur(img, (3, 3), 0)
kenel_55 = cv2.GaussianBlur(img, (5, 5), 0)
kenel_77 = cv2.GaussianBlur(img, (7, 7), 0)

cv2.imshow('img', img)
cv2.imshow('kenel_33', kenel_33)
cv2.imshow('kenel_55', kenel_55)
cv2.imshow('kenel_77', kenel_77)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 표준편차에 따른 블러

```python
img = cv2.imread('img.jpg')

sigma_1 = cv2.GaussianBlur(img, (0, 0), 1)  # sigmaX 가우시안 블러의 X축 방향의 표준편차
sigma_2 = cv2.GaussianBlur(img, (0, 0), 2)
sigma_3 = cv2.GaussianBlur(img, (0, 0), 3)

cv2.imshow('img', img)
cv2.imshow('sigma_1', sigma_1)
cv2.imshow('sigma_2', sigma_2)
cv2.imshow('sigma_3', sigma_3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 다림질 효과 (원근감)

```python
import numpy as np

img = cv2.imread('news.jpg')

width, height = 640, 200

src = np.array([[508,381],[1020, 371],[1120, 587],[457, 600]], dtype=np.float32)
dst = np.array([[0,0],[width, 0],[width, height],[0, height]], dtype=np.float32)
# 4개 꼭지점은 시계방향으로 좌표값 입력

matrix = cv2.getPerspectiveTransform(src, dst)
res = cv2.warpPerspective(img, matrix , (width, height))

cv2.imshow('img', img)
cv2.imshow('res', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 회전된 이미지를 정면에서 보는 효과

```python
import cv2
import numpy as np

img = cv2.imread('card.jpg')

# print(img)
width, height = 320, 450

src = np.array([[378, 257],[658, 109],[989, 372],[682, 583]], dtype=np.float32)
dst = np.array([[0,0],[width, 0],[width, height],[0, height]], dtype=np.float32)
# 4개 꼭지점은 시계방향으로 좌표값 입력

matrix = cv2.getPerspectiveTransform(src, dst)
res = cv2.warpPerspective(img, matrix , (width, height))

cv2.imshow('img', img)
cv2.imshow('res', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 마우스 이벤트

```python
import cv2
import numpy as np
import math

point4 = []
src_img = cv2.imread('card.jpg')
COLOR = (0, 0, 255)
def mouse_handler(event, x, y, flag, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        point4.append((x,y))
        if len(point4) == 4:
            show_result()
    
        for point in point4:
            cv2.circle(src_img, point , 10 , COLOR, cv2.FILLED )
    
        cv2.imshow('img', src_img)
  
def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def show_result():

    dists = []
    for i in range(4):
        d = distance(point4[i], point4[(i+1) % 4])
        dists.append(d)
    
    # 가로, 세로 구하기
    height = int(max(dists))
    width = int(min(dists))

    src = np.float32(point4)
    dst = np.array([[0,0],[width, 0],[width, height],[0, height]], dtype=np.float32)
    # 4개 꼭지점은 시계방향으로 좌표값 입력
    
    matrix = cv2.getPerspectiveTransform(src, dst)
    res = cv2.warpPerspective(src_img, matrix , (width, height))
    cv2.imshow('result', res)

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_handler)

cv2.imshow('img', src_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 마우스 이벤트

```python
import cv2
import numpy as np

point4 = []
img = cv2.imread('card.jpg')
color = (0,100,255)

def mouse_handler(event, x, y, flag, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        point4.append((x,y))
        print(x,y)

        if len(point4) == 4:
            show_result()
    
        for point in point4:
            cv2.circle(img, point, 5, color, cv2.FILLED)
            cv2.imshow('img', img)


def show_result():
    # 1. 찍은 점들 사이의 거리를 계산해서 실제 비율 찾기
    # 좌표 순서: 0:좌상, 1:우상, 2:우하, 3:좌하
    
    # 가로 길이 계산 (상단 두 점 사이 거리와 하단 두 점 사이 거리 중 큰 값)
    width_top = np.sqrt(((point4[1][0] - point4[0][0]) ** 2) + ((point4[1][1] - point4[0][1]) ** 2))
    width_bottom = np.sqrt(((point4[2][0] - point4[3][0]) ** 2) + ((point4[2][1] - point4[3][1]) ** 2))
    max_width = int(max(width_top, width_bottom))

    # 세로 길이 계산 (좌측 두 점 사이 거리와 우측 두 점 사이 거리 중 큰 값)
    height_left = np.sqrt(((point4[3][0] - point4[0][0]) ** 2) + ((point4[3][1] - point4[0][1]) ** 2))
    height_right = np.sqrt(((point4[2][0] - point4[1][0]) ** 2) + ((point4[2][1] - point4[1][1]) ** 2))
    max_height = int(max(height_left, height_right))

    # 2. 계산된 최대 길이를 바탕으로 dst 좌표 설정
    src = np.array(point4, dtype=np.float32)
    dst = np.array([
        [0, 0],
        [max_width, 0],
        [max_width, max_height],
        [0, max_height]
    ], dtype=np.float32)

    # 3. 변환 및 출력
    matrix = cv2.getPerspectiveTransform(src, dst)
    # 깨끗한 이미지를 위해 원본(origin)을 사용하도록 추천드려요! (이전 답변 참고)
    res = cv2.warpPerspective(img, matrix, (max_width, max_height), flags=cv2.INTER_LANCZOS4)

    cv2.imshow('Result', res)
    print(f"변환 완료! 사이즈: {max_width} x {max_height}")
    
    

cv2.namedWindow('img')
cv2.setMouseCallback('img', mouse_handler)

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

🛠️ 추가된 기능 설명
드래그 수정: 4개의 점을 다 찍은 후, 마우스로 점을 잡고 끌면 실시간으로 사각형 모양이 변합니다.

시각적 가이드: 점 번호(1~4)와 연결선(초록색)을 그려서 어떤 순서로 변환될지 미리 보여줍니다.

인터랙션:

Enter: 현재 사각형 영역을 펴서 Result 창에 띄웁니다.

R (Reset): 점을 다 지우고 처음부터 다시 찍습니다.

ESC: 프로그램 종료.

```python
import cv2
import numpy as np

# 설정값
WIDTH, HEIGHT = 650, 250
RADIUS = 10  # 점의 크기 (클릭 감지 범위)

src_pts = []
drag_idx = -1  # 드래그 중인 점의 인덱스 (-1은 없음)

def mouse_handler(event, x, y, flags, param):
    global src_pts, drag_idx, img_display

    # 1. 점 찍기 (4개 미만일 때)
    if event == cv2.EVENT_LBUTTONDOWN and len(src_pts) < 4:
        src_pts.append([x, y])
        draw_and_show()

    # 2. 드래그 시작 (4개 다 찍힌 상태에서 점을 클릭했을 때)
    elif event == cv2.EVENT_LBUTTONDOWN and len(src_pts) == 4:
        for i, pt in enumerate(src_pts):
            if np.linalg.norm(np.array(pt) - [x, y]) < RADIUS:
                drag_idx = i
                break

    # 3. 드래그 중
    elif event == cv2.EVENT_MOUSEMOVE and drag_idx != -1:
        src_pts[drag_idx] = [x, y]
        draw_and_show()

    # 4. 드래그 종료
    elif event == cv2.EVENT_LBUTTONUP:
        drag_idx = -1

def draw_and_show():
    img_draw = origin_img.copy()
    
    # 점과 선 그리기
    if len(src_pts) > 0:
        for i, pt in enumerate(src_pts):
            cv2.circle(img_draw, tuple(pt), RADIUS, (0, 0, 255), -1)
            cv2.putText(img_draw, str(i+1), (pt[0]+10, pt[1]-10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)
        
        # 4개가 찍히면 사각형 연결
        if len(src_pts) == 4:
            pts_array = np.array(src_pts, np.int32)
            cv2.polylines(img_draw, [pts_array], True, (0, 255, 0), 2)
            cv2.putText(img_draw, "Press 'Enter' to Transform", (20, 30), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

    cv2.imshow('Select & Drag', img_draw)

def transform():
    if len(src_pts) != 4: return
    
    src = np.array(src_pts, dtype=np.float32)
    dst = np.array([[0, 0], [WIDTH, 0], [WIDTH, HEIGHT], [0, HEIGHT]], dtype=np.float32)

    matrix = cv2.getPerspectiveTransform(src, dst)
    res = cv2.warpPerspective(origin_img, matrix, (WIDTH, HEIGHT), flags=cv2.INTER_LANCZOS4)
    cv2.imshow('Result', res)

# 실행부
origin_img = cv2.imread('news.jpg')
if origin_img is None:
    print("이미지를 불러올 수 없습니다.")
    exit()

cv2.imshow('Select & Drag', origin_img)
cv2.setMouseCallback('Select & Drag', mouse_handler)

print("1. 4개의 점을 시계방향으로 찍으세요.")
print("2. 점을 드래그해서 위치를 수정하세요.")
print("3. 'Enter'를 누르면 변환됩니다. 'r'을 누르면 초기화됩니다.")

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 13:  # Enter 키
        transform()
    elif key == ord('r'):  # r 키로 초기화
        src_pts = []
        cv2.imshow('Select & Drag', origin_img)
    elif key == 27:  # ESC 키로 종료
        break

cv2.destroyAllWindows()
```

### 사이즈 자동계산 버전

```python
import cv2
import numpy as np

# 설정값
WIDTH, HEIGHT = 650, 250
RADIUS = 5  # 점의 크기 (클릭 감지 범위)

src_pts = []
drag_idx = -1  # 드래그 중인 점의 인덱스 (-1은 없음)

def mouse_handler(event, x, y, flags, param):
    global src_pts, drag_idx, img_display

    # 1. 점 찍기 (4개 미만일 때)
    if event == cv2.EVENT_LBUTTONDOWN and len(src_pts) < 4:
        src_pts.append([x, y])
        draw_and_show()

    # 2. 드래그 시작 (4개 다 찍힌 상태에서 점을 클릭했을 때)
    elif event == cv2.EVENT_LBUTTONDOWN and len(src_pts) == 4:
        for i, pt in enumerate(src_pts):
            if np.linalg.norm(np.array(pt) - [x, y]) < RADIUS:
                drag_idx = i
                break

    # 3. 드래그 중
    elif event == cv2.EVENT_MOUSEMOVE and drag_idx != -1:
        src_pts[drag_idx] = [x, y]
        draw_and_show()

    # 4. 드래그 종료
    elif event == cv2.EVENT_LBUTTONUP:
        drag_idx = -1

def draw_and_show():
    img_draw = origin_img.copy()
    
    # 점과 선 그리기
    if len(src_pts) > 0:
        for i, pt in enumerate(src_pts):
            cv2.circle(img_draw, tuple(pt), RADIUS, (0, 0, 255), -1)
            cv2.putText(img_draw, str(i+1), (pt[0]+10, pt[1]-10), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)
        
        # 4개가 찍히면 사각형 연결
        if len(src_pts) == 4:
            pts_array = np.array(src_pts, np.int32)
            cv2.polylines(img_draw, [pts_array], True, (0, 255, 0), 2)
            cv2.putText(img_draw, "Press 'Enter' to Transform", (20, 30), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

    cv2.imshow('Select & Drag', img_draw)

def transform():
    if len(src_pts) != 4: return
    
    # 1. 찍은 점들 사이의 거리 계산 (피타고라스 정리)
    # src_pts 순서: 0:좌상, 1:우상, 2:우하, 3:좌하 (시계방향 전제)
    s = np.array(src_pts, dtype=np.float32)

    # 가로 길이 계산 (위쪽 두 점 거리 vs 아래쪽 두 점 거리 중 큰 값)
    w1 = np.sqrt(((s[1][0] - s[0][0])**2) + ((s[1][1] - s[0][1])**2))
    w2 = np.sqrt(((s[2][0] - s[3][0])**2) + ((s[2][1] - s[3][1])**2))
    max_w = int(max(w1, w2))

    # 세로 길이 계산 (왼쪽 두 점 거리 vs 오른쪽 두 점 거리 중 큰 값)
    h1 = np.sqrt(((s[3][0] - s[0][0])**2) + ((s[3][1] - s[0][1])**2))
    h2 = np.sqrt(((s[2][0] - s[1][0])**2) + ((s[2][1] - s[1][1])**2))
    max_h = int(max(h1, h2))

    # 2. 자동 계산된 크기로 목적지(dst) 좌표 설정
    src = s
    dst = np.array([
        [0, 0],
        [max_w, 0],
        [max_w, max_h],
        [0, max_h]
    ], dtype=np.float32)

    # 3. 고화질 변환 (max_w, max_h 적용)
    matrix = cv2.getPerspectiveTransform(src, dst)
    res = cv2.warpPerspective(origin_img, matrix, (max_w, max_h), flags=cv2.INTER_CUBIC)

    # 4. 샤프닝 필터 적용
    blurred = cv2.GaussianBlur(res, (0, 0), 3)
    res_sharp = cv2.addWeighted(res, 1.5, blurred, -0.5, 0)

    # 결과 출력
    cv2.imshow('Standard (Cubic)', res)
    cv2.imshow('Sharpened (High Contrast)', res_sharp)
    print(f"변환 완료! 최적 사이즈 계산됨: {max_w} x {max_h}")

# 실행부
origin_img = cv2.imread('news.jpg')
if origin_img is None:
    print("이미지를 불러올 수 없습니다.")
    exit()

cv2.imshow('Select & Drag', origin_img)
cv2.setMouseCallback('Select & Drag', mouse_handler)

print("1. 4개의 점을 시계방향으로 찍으세요.")
print("2. 점을 드래그해서 위치를 수정하세요.")
print("3. 'Enter'를 누르면 변환됩니다. 'r'을 누르면 초기화됩니다.")

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 13:  # Enter 키
        transform()
    elif key == ord('r'):  # r 키로 초기화
        src_pts = []
        cv2.imshow('Select & Drag', origin_img)
    elif key == 27:  # ESC 키로 종료
        break

cv2.destroyAllWindows()
```

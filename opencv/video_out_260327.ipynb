### OpenCV를 이용하여 세로 동영상을 가로 동영상으로 회전
- pexels.com 사이트에서 적당한 동영상 다운로드 후 작업
- 시계 반대방향으로 90도 회전
- 재생 속도는 원본의 2배
- 코덱 : DIVX
- 출력파일 : video_out.avi

```python
import cv2

cap = cv2.VideoCapture('cat.mp4')

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
print(width, height, fps)

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('video_out.avi', fourcc, fps*2, (height,width))

while cap.isOpened():
    ret, frame = cap.read()
    # 프레임을 제대로 못 읽어왔다면 루프 종료
    if not ret:
        print("프레임을 수신할 수 없습니다. 종료합니다.")
        break  
        
    frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)
    out.write(frame)
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2

# 1. 파일 불러오기
cap = cv2.VideoCapture('cat.mp4')

# 원본 동영상의 정보 가져오기 (너비, 높이, FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# 2. 재생 속도 2배 설정 (FPS를 2배로 설정)
double_fps = fps * 2

# 3. 90도 회전 시 너비와 높이가 서로 바뀜을 주의!
# 가로 영상으로 출력하려면 출력 사이즈를 (height, width) 순서로 바꿔야 합니다.
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('video_out.avi', fourcc, double_fps, (height, width))

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print("프레임을 수신할 수 없거나 영상이 끝났습니다.")
        break
    
    # 4. 시계 반대방향으로 90도 회전
    # ROTATE_90_COUNTERCLOCKWISE 또는 ROTATE_90_CLOCKWISE 후 뒤집기 등을 사용
    frame = cv2.rotate(frame, cv2.ROTATE_90_COUNTERCLOCKWISE)

    # 결과 파일 저장
    out.write(frame)

    # 화면 표시
    cv2.imshow('frame', frame)

    # 5. 대기 시간 조절 (재생 속도 확인용)
    # waitKey(0)은 키를 누를 때까지 멈추므로, 적절한 숫자를 넣어야 영상처럼 보입니다.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 리소스 해제
cap.release()
out.release() # out 객체도 반드시 닫아주어야 파일이 정상 저장됩니다.
cv2.destroyAllWindows()
```

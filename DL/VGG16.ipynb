```python
import keras
from keras import layers
```

```python
vggnet = keras.Sequential()
vggnet.add(layers.Input(shape=(224, 224, 3)))

for n in [64, 128]:
    for _ in range(2):
        vggnet.add(layers.Conv2D(filters=n, kernel_size=3, padding='same', activation='relu'))
    vggnet.add(layers.MaxPooling2D(2))

for n in [256, 512, 512]:
    for _ in range(3):
        vggnet.add(layers.Conv2D(filters=n, kernel_size=3, padding='same',activation='relu'))
    vggnet.add(layers.MaxPooling2D(2))

vggnet.add(layers.Flatten())
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(1000, activation='softmax'))
```

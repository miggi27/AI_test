### VGGNet : 2014년 준우승, 기존 CNN과의 차별, 반복구조
- input(224*224*3) -> Conv2D - 64, 3, same, relu
- -> Conv2D - 64, 3, same, relu
- -> polling - 2
- -> Conv2D - 128, 3, same, relu
- -> Conv2D - 128, 3, same, relu
- -> polling - 2
- -> Conv2D - 256, 3, same, relu
- -> Conv2D - 256, 3, same, relu
- -> Conv2D - 256, 3, same, relu
- -> polling - 2
- -> Conv2D - 512, 3, same, relu
- -> Conv2D - 512, 3, same, relu
- -> Conv2D - 512, 3, same, relu
- -> polling - 2
- -> Conv2D - 512, 3, same, relu
- -> Conv2D - 512, 3, same, relu
- -> Conv2D - 512, 3, same, relu
- -> polling - 2
- -> Flatten
- -> Dense(4096, relu)
- -> Dense(4096, relu)
- -> Dense(1000, softmax)

```python
import keras
from keras import layers
```

```python
VGGNet = keras.Sequential()
VGGNet.add(layers.Input(shape=(224,224,3)))
VGGNet.add(layers.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.MaxPooling2D(pool_size=3, strides=2))

VGGNet.add(layers.Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.MaxPooling2D(pool_size=3, strides=2))

VGGNet.add(layers.Conv2D(filters=256, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=256, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=256, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.MaxPooling2D(pool_size=3, strides=2))

VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.MaxPooling2D(pool_size=3, strides=2))

VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu'))
VGGNet.add(layers.MaxPooling2D(pool_size=3, strides=2))

VGGNet.add(layers.Flatten())
VGGNet.add(layers.Dense(4096, activation='relu'))
VGGNet.add(layers.Dense(4096, activation='relu'))
VGGNet.add(layers.Dense(1000, activation='softmax'))
```

```python
vggnet = keras.Sequential()
vggnet.add(layers.Input(shape=(224,224,3)))
# (필터 수, 반복 횟수)
vgg_config = [
    (64, 2), (128, 2),  # 앞선 2개 블록
    (256, 3), (512, 3), (512, 3) # 뒤의 3개 블록
]

for n_filters, n_layers in vgg_config:
    for _ in range(n_layers):
        vggnet.add(layers.Conv2D(filters=n_filters, kernel_size=3, padding='same', activation='relu'))
    vggnet.add(layers.MaxPooling2D(2))

vggnet.add(layers.Flatten())
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(4096, activation='relu'))
vggnet.add(layers.Dense(1000, activation='softmax'))
```

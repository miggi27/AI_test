- Adaptive Threshold

```python
import cv2
img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

def evt(val):
    # print(val)
    pass

name = 'Trackbar'
cv2.namedWindow(name)

cv2.createTrackbar('block_size', name, 25, 100 , evt) 
# block_size의 값은 1보다 큰 홀수
cv2.createTrackbar('c', name, 3, 10 , evt)

while True:
    block_size =cv2.getTrackbarPos('block_size', name)
    c =cv2.getTrackbarPos('c', name)

    if block_size <= 1:
        block_size = 3

    if block_size % 2 == 0:
        block_size += 1
   
    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, 
                                        cv2.THRESH_BINARY, block_size, c)
    
    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q'):
        break
    
cv2.destroyAllWindows()
```

- 오츠(otsu) : 최적의 임계값을 찾아주는 알고리즘

```python
import cv2
img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img , 127 , 255, cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(img , -1 , 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print('otsu : ', ret)

cv2.imshow('img', img)
cv2.imshow('binary', binary)
cv2.imshow('otsu', otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 노이즈 제거
- dilate(팽창) : 이미지를 확장해서 작은 노이즈를 제거

```python
import numpy as np

kernel = np.ones((3,3), dtype = np.uint8)
kernel
```

```python
import cv2
img = cv2.imread('erode.png', cv2.IMREAD_GRAYSCALE)
dilate1 = cv2.dilate(img, kernel, iterations=1)
dilate2 = cv2.dilate(img, kernel, iterations=2)
dilate3 = cv2.dilate(img, kernel, iterations=3)

cv2.imshow('img', img)
cv2.imshow('dilate1', dilate1)
cv2.imshow('dilate2', dilate2)
cv2.imshow('dilate3', dilate3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE)
erode1 = cv2.erode(img, kernel, iterations=1)
erode2 = cv2.erode(img, kernel, iterations=2)
erode3 = cv2.erode(img, kernel, iterations=3)

cv2.imshow('img', img)
cv2.imshow('erode1', erode1)
cv2.imshow('erode2', erode2)
cv2.imshow('erode3', erode3)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

- 이미지변환 (열림 & 닫힘)
- 열림 : 침식 후 팽창. 노이즈를 깎아서 제거 한 후에 팽창
- 열림 : dilate(erode(image))

```python
import cv2
img = cv2.imread('dilate.png', cv2.IMREAD_GRAYSCALE)
erode = cv2.erode(img, kernel, iterations=1)
dilate = cv2.dilate(erode, kernel, iterations=1)


cv2.imshow('img', img)
cv2.imshow('erode', erode)
cv2.imshow('dilate', dilate)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img = cv2.imread('erode.png', cv2.IMREAD_GRAYSCALE)
dilate = cv2.dilate(img, kernel, iterations=2)
erode = cv2.erode(dilate, kernel, iterations=2)

cv2.imshow('img', img)
cv2.imshow('erode', erode)
cv2.imshow('dilate', dilate)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 이미지 검출 - 윤곽선
- 윤곽선(contour)

```python
img = cv2.imread('card5.png')
tar_img = img.copy()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, otsu = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

COLOR = (0, 0, 250)

contours, hiera = cv2.findContours(otsu, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

for cont in contours:
    x, y, w, h = cv2.boundingRect(cont)
    cv2.rectangle(tar_img , (x,y), (x+w, y+h), COLOR, 2)

# cv2.drawContours(tar_img, contours, -1, COLOR, 2)


cv2.imshow('img', img)
cv2.imshow('gray', gray)
cv2.imshow('otsu', otsu)
cv2.imshow('contours', tar_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('card5.png')
tar_img = img.copy()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, otsu = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

COLOR = (0, 250, 250)

contours, hiera = cv2.findContours(otsu, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for cnt in contours:
    if cv2.contourArea(cnt) > 25000:
        x, y, w, h = cv2.boundingRect(cnt)
        cv2.rectangle(tar_img, (x, y), (x + w, y+ h), COLOR, 2)

cv2.imshow('img', img)
cv2.imshow('gray', gray)
cv2.imshow('otsu', otsu)
cv2.imshow('contours', tar_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 프레임 생성
data = {
    '이름':['김길동','이길동','강길동','홍길동','박길동','나길동'],
    '주소':['광주','서울','대구','대전','광주','서울'],
    '나이':[33,23,30,20,21,25],
    '점수':[76,88,98,100,60,90],
    '언어':['Python','Java','C++','','','C']
}
df = pd.DataFrame(data)

# 2. 한글 폰트 설정 (그래프 깨짐 방지)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# 3. 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 왼쪽 축 (ax1): 점수를 막대 그래프로
ax1.bar(df['이름'], df['점수'], color='skyblue', label='점수', alpha=0.5)
ax1.set_ylabel('점수 (points)')
ax1.set_ylim(0, 110) # 점수 범위 설정
ax1.legend(loc='upper left')

# 오른쪽 축 (ax2): 나이를 선 그래프로 (이중 축 사용)
ax2 = ax1.twinx() 
ax2.plot(df['이름'], df['나이'], color='red', marker='o', ls='--', label='나이', linewidth=2)
ax2.set_ylabel('나이 (age)')
ax2.set_ylim(0, 40) # 나이 범위 설정
ax2.legend(loc='upper right')

# 제목 설정
plt.title('길동이들의 점수 및 나이 분포')
plt.show()
```

### 💡 그래프 포인트 설명!
- ax1.bar: 점수는 막대 그래프로 그려서 전체적인 크기를 비교하기 좋게 했습니다.
- ax1.twinx(): 이 코드가 핵심입니다! ax1과 X축(이름)은 공유하지만, Y축만 하나 더 오른쪽에 새로 만들겠다는 뜻입니다.
- ax2.plot: 나이는 선 그래프로 그려서 점수 막대와 겹치지 않고 추이를 볼 수 있게 했습니다.
- alpha=0.7: 막대를 살짝 투명하게 해서 뒤에 선이 가려지지 않게 조절했습니다.
---
- fig (Figure): 도화지 전체를 말합니다. 그래프가 그려질 커다란 창(윈도우) 자체예요. 나중에 그림을 파일로 저장(fig.savefig('graph.png'))할 때 이 변수를 사용합니다.
- ax1 (Axes): 도화지 안에 들어가는 실제 그래프 한 칸입니다. 축(Axis), 제목, 데이터 등이 그려지는 실질적인 '그림 영역'이죠.
- subplots: "도화지(fig)랑 그림 칸(ax)을 동시에 만들어줘!"라는 명령어입니다.
- twinx(): "이미 있는 ax1과 X축은 똑같이 쓰고(twin), **Y축만 오른쪽(x의 반대편)**에 하나 더 복제해줘!"라는 뜻입니다.
- 이때 비로소 새로운 Y축을 가진 **ax2**라는 새로운 그림 칸이 탄생해서 변수에 담긴 겁니다.
- 만약 ax2를 만들지 않고 ax1에 '나이'를 그냥 그렸다면, 왼쪽의 0~100 눈금 하나로 점수와 나이를 같이 표현해야 해서 그래프가 엉망이 되었을 거예요.

```python
# 1. 데이터 프레임 생성
data = {
    '이름':['김길동','이길동','강길동','홍길동','박길동','나길동'],
    '주소':['광주','서울','대구','대전','광주','서울'],
    '나이':[33,23,30,20,21,25],
    '국어':[76, 88, 98, 100, 60, 90],
    '영어':[85, 92, 70, 88, 95, 80],
    '수학':[90, 80, 85, 95, 70, 100],
    '언어':['Python','Java','C++','','','C']
}
df = pd.DataFrame(data)

# 2. 한글 폰트 설정 (그래프 깨짐 방지)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

# 3. 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 왼쪽 축 (ax1): 이제 나이를 막대로 그립니다.
ax1.bar(df['이름'], df['나이'], color='lightcoral', label='나이', alpha=0.6)
ax1.set_ylabel('나이 (age)')
ax1.set_ylim(0, 50) # 나이 눈금에 맞게 조정
ax1.legend(loc='upper left')

# 데이터에 국어, 영어, 수학이 있다고 가정할 때 ax2 부분만 이렇게 늘어납니다:
ax2 = ax1.twinx()
ax2.plot(df['이름'], df['국어'], marker='o', label='국어')
ax2.plot(df['이름'], df['영어'], marker='^', label='영어')
ax2.plot(df['이름'], df['수학'], marker='s', label='수학')
ax2.set_ylabel('과목별 점수')
ax2.legend(loc='upper right')

plt.title('길동이들의 나이(막대) 및 점수(선) 분포')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 준비 (과목 추가)
data = {
    '이름':['김길동','이길동','강길동','홍길동','박길동','나길동'],
    '국어':[76, 88, 98, 100, 60, 90],
    '영어':[85, 92, 70, 88, 95, 80],
    '수학':[90, 80, 85, 95, 70, 100]
}
df = pd.DataFrame(data)
df.set_index('이름', inplace=True) # 이름을 인덱스로 잡아야 그래프가 예쁘게 나와요

# 2. 한글 설정
plt.rc('font', family='Malgun Gothic')

# 3. 그래프 그리기 (Kind='bar' 사용)
ax = df.plot(kind='bar', figsize=(10, 6), width=0.8)
# ax = df.plot(kind='bar', figsize=(10, 6), stacked=True, alpha=0.8)
# df.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('과목별 성적 비교')
plt.ylabel('점수')
plt.xticks(rotation=0) # 이름 가로로 보기 좋게 유지
plt.legend(loc='upper right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 데이터 준비
data = {
    '이름':['김길동','이길동','강길동','홍길동','박길동','나길동'],
    '국어':[76, 88, 98, 100, 60, 90],
    '영어':[85, 92, 70, 88, 95, 80],
    '수학':[90, 80, 85, 95, 70, 100]
}
df = pd.DataFrame(data)

# 2. 한글 설정
plt.rc('font', family='Malgun Gothic')

# 3. 막대 위치 계산 (핵심!)
index = np.arange(len(df)) # [0, 1, 2, 3, 4, 5]
w = 0.25 # 막대 너비

# 4. 그래프 그리기
plt.figure(figsize=(10, 6))

# 각 과목의 막대를 너비(w)만큼 옆으로 밀어서 그립니다.
plt.bar(index - w, df['국어'], width=w, label='국어')
plt.bar(index,     df['영어'], width=w, label='영어')
plt.bar(index + w, df['수학'], width=w, label='수학')

# 5. 축 눈금 설정
plt.xticks(index, df['이름']) # 0, 1, 2... 위치를 이름으로 바꿈
plt.title('과목별 성적비교')
plt.legend()
plt.show()
```

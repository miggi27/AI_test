```python
import torch
import transformers

# 1. 일단 torch가 잘 로드되는지 다시 확인 (위에서 확인했으니 통과)
print(f"이미 설치된 Torch 버전: {torch.__version__}")

# 2. 핵심: transformers가 torch를 인식하지 못할 때 강제로 인식시키는 설정
from transformers.utils import import_utils
import_utils._torch_available = True  # "응, torch 있어!"라고 강제로 말해줌

# 3. 이제 pipeline 실행
from transformers import pipeline

try:
    # 다시 한번 시도
    classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
    result = classifier("I love studying Pandas with Gemini!")
    print("\n[성공 결과]:", result)
except Exception as e:
    print("\n[또 다른 에러 발생]:", e)
```

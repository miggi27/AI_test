#### DataFrame 만들 딕셔너리 만들기

```python
data = {
    '이름':['김길동','이길동','강길동','홍길동','박길동','나길동'],
    '주소':['광주','서울','대구','대전','광주','서울'],
    '나이':[33,23,30,20,21,25],
    '점수':[76,88,98,100,60,90],
    '언어':['Python','Java','C++','','','C']
}
data
```

### pandas import

```python
import pandas as pd
```

### DataFrame 생성

```python
df=pd.DataFrame(data)
df
```

### DataFrame index 설정

```python
df.index = ['1번','2번','3번','4번','5번','6번']
df
```

```python
df.index = [f"{i+1}번" for i in range(len(df))] # 대량의 데이터에서는 이 방법이 좋겠네
df
```

```python
# 인덱스를 1번, 2번... 형태로 변경
# df.index = [f"{i+1}번" for i in range(len(df))]
df = pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번'])
df
```

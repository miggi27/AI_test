```python
import matplotlib.pyplot as plt
```

```python
x = [1,3,5]
y = [2,4,9]
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show()
```

- 제목 설정

```python
plt.plot(x, y)
plt.title('Line Graph')
plt.show()
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
plt.show()
```

- 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] =  15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
plt.show()
```

```python
plt.plot([-1, 1, 4], [-3, -1, 5])
plt.title('꺾은선 그래프')
plt.show()
```

#### 축 설정

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
plt.show()
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
plt.xlabel('X축', color = 'red')
plt.ylabel('Y축', color = '#00ff00')
plt.show()
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
plt.xlabel('X축', color = 'red', loc='right')
plt.ylabel('Y축', color = '#00ff00', loc='top')
plt.show()
```

```python
plt.plot(x, y)
plt.xticks([1,3,5])
plt.yticks([2,4, 6, 8])
plt.show()
```

```python
plt.plot(x, y)
plt.xlim(1,10)
plt.ylim(2,15)
plt.show()
```

#### 범례(legend)

```python
plt.plot(x, y, label='점수')
plt.legend()
```

```python
plt.plot(x, y, label='점수')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label='점수')
plt.legend(loc=7)
```

```python
plt.plot(x, y, marker='o', linestyle='None')
plt.show()
```

```python
plt.plot(x, y, marker='v', markersize=10)
plt.show()
```

```python
plt.plot(x, y, marker='X', markersize=10)
plt.show()
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='red')
plt.show()
```

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='red', markerfacecolor='yellow')
plt.show()
```

#### 포맷

```python
plt.plot(x, y, 'ro--')  # color marker linestyle
plt.show()
```

```python
plt.plot(x, y, 'bv:')  # color marker linestyle
plt.show()
```

```python
plt.plot(x, y, 'go')  # color marker linestyle
plt.show()
```

#### 축약어

```python
plt.plot(x, y, marker='X', markersize=10, markeredgecolor='red')
plt.show()
```

```python
plt.plot(x, y, marker='X', markersize=10, mfc='red')
plt.show()
```

```python
plt.plot(x, y, marker='X', ms=10, mfc='g', mec='r')
plt.show()
```

#### 투명도

```python
plt.plot(x, y, marker='X', ms=10, mfc='r', alpha=0.5)
plt.show()
```

#### 그래프 크기

```python
plt.figure(figsize=(8,4))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,4), dpi=50)
plt.plot(x,y)
plt.show()
```

#### 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
plt.show()
```

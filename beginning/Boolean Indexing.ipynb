#### str 함수

##### 문자열 데이터에서 시작 문자 검색

```python
cond = df['이름'].str.startswith('박')
df[cond]
```

```python
cond = df['이름'].str.startswith('박') # '박'으로 시작하는 것들만 출
df[cond]
```

##### 문자열 데이터에서 포함 문자 검색

```python
cond = df['주소'].str.contains('주') # '주'자가 포함된 것들만 출력
df[cond]
```

```python
df[~cond]
```

```python
lang = ['Java','Python']
cond = df['언어'].isin(lang)
df[cond]
```

```python
df[~cond]
```

```python
lang = ['java','python']
cond = df['언어'].isin(lang)
df[cond]
```

```python
df['언어'].str.lower()
```

```python
lang = ['java','python']
cond = df['언어'].str.lower().isin(lang)
df[cond]
```

```python
cond = df['언어'].str.contains('Java')
df[cond]
```

```python
cond = df['언어'].str.contains('Java', na=False)
df[cond]
```

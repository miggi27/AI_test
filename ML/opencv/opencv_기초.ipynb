```python
import cv2
import os

# 1. 파일이 진짜 그 자리에 있는지 먼저 확인!
file_path = 'people.jpg'

if not os.path.exists(file_path):
    print(f"❌ 파일을 찾을 수 없습니다: {os.path.abspath(file_path)}")
    print("현재 작업 폴더 위치:", os.getcwd()) # 현재 어디서 작업 중인지 출력
else:
    # 2. 존재한다면 읽기 시도
    image = cv2.imread(file_path)
    
    if image is None:
        print("❌ 파일은 있는데 OpenCV가 읽지 못합니다. (한글 경로 혹은 파일 손상)")
    else:
        print(f"✅ 성공! 이미지 크기: {image.shape}")
        cv2.imshow('Original', image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```

```python
gray_color = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray', gray_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Canny => edge 경계선을 따는 툴
edges = cv2.Canny(gray_color, 50, 150)
cv2.imshow('edges', edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

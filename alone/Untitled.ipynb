## 🔍 1단계: 데이터 분석 (EDA) - "데이터와 통성명하기"
*데이터를 받자마자 "이 녀석의 정체가 뭐지?"를 파악하는 단계입니다.*

### 1) 데이터 로드 및 훑어보기
* `df = pd.read_csv('data.csv')` : 데이터를 일단 가져온다.
* `df.head()` : 상위 5개 행을 보며 데이터의 생김새를 익힌다.
* `df.tail()` : 하위 5개 행을 확인한다.

### 2) 데이터 상태 진단 (필수 명령어 5개)
| 명령어 | 용도 (의사의 질문) | 체크포인트 |
| :--- | :--- | :--- |
| **df.info()** | "전체적인 체격이 어때?" | 데이터 타입(int, float, object), 전체 행 개수 확인 |
| **df.isnull().sum()** | "어디 아픈(비어있는) 곳 있어?" | **결측치(NaN)** 개수 파악 (전처리의 시작점) |
| **df.describe()** | "수치들이 정상 범위야?" | 평균, 최솟값, 최댓값, 이상치(Outlier) 존재 확인 |
| **df.shape** | "덩치가 얼마나 커?" | 전체 데이터의 (행, 열) 개수 확인 |
| **df['컬럼'].value_counts()** | "항목별로 골고루 있어?" | 범주형 데이터의 종류와 빈도수 확인 |



---

## 🛠 2단계: 데이터 전처리 (Preprocessing) - "본격적인 손질"
*모델(AI)이 공부하기 좋게 데이터를 깔끔하게 수술하는 단계입니다.*

### 1) 결측치 처리 (Missing Values)
* **도려내기(삭제)**: `df.dropna(axis=0)` (빈칸이 있는 행을 통째로 삭제)
* **메우기(채우기)**: `df['컬럼'].fillna(df['컬럼'].mean(), inplace=True)` (평균이나 0으로 채움)

### 2) 데이터 형태 변환
* **타입 변경**: `df['열'].astype('int')` (계산이 필요한데 '문자'로 되어 있을 때 변경)
* **중복 제거**: `df.drop_duplicates()` (중복된 데이터는 학습을 방해하므로 제거)

### 3) 범주형 인코딩 (Encoding)
*AI는 글자를 못 읽습니다. 무조건 숫자로 바꿔야 합니다.*
* **원-핫 인코딩(One-Hot)**: `df = pd.get_dummies(df, columns=['컬럼명'])`
    * *예: [빨강, 파랑] → [1, 0], [0, 1]로 변신!*



### 4) 수치 스케일링 (Scaling)
*값의 범위가 너무 다르면 모델이 오해합니다. (예: 가격 10억 vs 평점 4.5)*
* **StandardScaler**: 평균을 0, 표준편차를 1로 맞춤 (종 모양 분포)
* **MinMaxScaler**: 모든 값을 0과 1 사이로 압축 (단위 일치)

### 1) 데이터 로드 및 훑어보기
* `df = pd.read_csv('data.csv')` : 데이터를 메모리로 호출
* `df.head()` : 상위 데이터 5줄 확인 (데이터 관찰)
* `df.tail()` : 하위 데이터 5줄 확인

### 2) 데이터 상태 진단 (필수 명령어)
| 명령어 | 용도 (의사의 질문) | 체크포인트 |
| :--- | :--- | :--- |
| **df.info()** | "전체적인 체격이 어때?" | 데이터 타입(숫자/문자), 전체 행 개수 확인 |
| **df.isnull().sum()** | "어디 빈 곳(결측치)이 있어?" | **결측치(NaN)** 개수 파악 (가장 중요) |
| **df.describe()** | "수치들이 정상 범위야?" | 평균, 최솟값, 최댓값, 이상치 확인 |

## 🛠 2단계: 데이터 전처리 (Preprocessing) - "치료 단계"
*모델(AI)이 학습할 수 있도록 데이터를 깨끗하고 균일하게 가공합니다.*

### 1) 결측치 처리 (Missing Values)
* **삭제**: `df.dropna(axis=0)` (빈칸 있는 행 제거)
* **채우기**: `df['컬럼'].fillna(df['컬럼'].mean(), inplace=True)` (평균 등으로 보충)

### 2) 데이터 형태 변환
* **타입 변경**: `df['열'].astype('int')` (계산 가능하게 변환)
* **중복 제거**: `df.drop_duplicates()` (중복 학습 방지)

### 3) 범주형 인코딩 (Encoding)
*AI는 텍스트를 인식 못 하므로 숫자로 변환합니다.*
* **원-핫 인코딩(One-Hot)**: `df = pd.get_dummies(df, columns=['컬럼명'])`
    * *예: [사과, 배] → [1, 0], [0, 1]*

### 4) 수치 스케일링 (Scaling)
*값의 범위(Scale)가 다르면 모델이 편향되므로 단위를 맞춥니다.*
* **StandardScaler**: 평균 0, 표준편차 1로 변환
* **MinMaxScaler**: 모든 값을 0~1 사이로 압축

5. **단위맞추기**: `Scaling` (단위 통일)

````python

데이터 전처리란?
데이터 전처리란, 데이터 분석을 위해 수집한 데이터를 분석에 적합한 형태로 가공하는 과정입니다. 
데이터 전처리를 통해 불필요한 데이터를 제거하고, 결측치나 이상치를 처리하여 데이터의 질을 향상시킬 수 있습니다. 
이렇게 가공된 데이터는 분석 모델을 구축하고 결과를 도출하는 데에 더욱 유용하게 활용될 수 있습니다.

데이터 전처리의 중요성
데이터 분석은 모든 산업 분야에서 큰 역할을 하고 있습니다. 
하지만 분석에 사용되는 데이터의 질이 좋지 않다면, 분석 결과도 부정확하게 나올 수 있습니다. 
따라서 데이터 전처리는 정확한 분석 결과를 얻기 위해 반드시 필요한 작업입니다.

데이터 전처리 예시
데이터 전처리에는 다양한 기법이 사용됩니다. 예를 들어, 다음과 같은 전처리 작업이 필요할 수 있습니다.

결측치 처리: 데이터에서 빠진 값이 있을 경우, 해당 값을 대체하거나 삭제하여 데이터의 일관성을 유지합니다.
이상치 처리: 데이터에서 이상한 값이 있을 경우, 해당 값을 대체하거나 삭제하여 분석 결과에 영향을 미치는 오류를 방지합니다.
데이터 정규화: 서로 다른 스케일의 데이터를 비교 분석하기 위해, 데이터 값을 일정한 범위로 조정합니다.
데이터 인코딩: 텍스트 데이터를 컴퓨터가 이해할 수 있는 형태로 변환합니다. (예: 원-핫 인코딩)
데이터 통합: 여러 개의 데이터를 하나의 데이터로 통합하여 분석에 용이하게 합니다.
데이터 분할: 분석에 필요한 부분 데이터를 추출하여, 불필요한 데이터를 제거합니다.
데이터 정렬: 분석에 필요한 순서대로 데이터를 정렬합니다.
데이터 그룹화: 데이터를 그룹별로 분류하고, 각 그룹에 대한 통계 정보를 추출합니다. (예: 그룹별 평균, 합계, 표준편차 등)
데이터 변환 함수: apply, map, applymap 등의 함수를 사용하여 데이터 값을 변환합니다.
데이터 피벗: pivot_table 함수를 사용하여, 행과 열을 바꾸거나, 그룹별 집계 정보를 표현합니다.
데이터 병합: merge 함수를 사용하여, 여러 개의 데이터를 하나로 병합합니다.
데이터 분할: split 함수를 사용하여, 데이터를 분할하고, 분할된 데이터를 분석합니다.
데이터 샘플링: sample 함수를 사용하여, 샘플 데이터를 추출하고, 추출된 데이터를 분석합니다.
데이터 집계: agg 함수를 사용하여, 그룹별 집계 정보를 추출합니다.
데이터 시각화: matplotlib, seaborn 등의 시각화 라이브러리를 사용하여, 
데이터를 시각화하고, 분석 결과를 쉽게 이해할 수 있도록 도와줍니다.

데이터 전처리 과정
데이터 전처리 과정은 다음과 같이 진행됩니다.

데이터 수집
데이터 정제 (결측치, 이상치 처리)
데이터 변환 (날짜, 문자열 등)
데이터 필터링 (조건에 따른 데이터 추출)
데이터 정렬 (sort_values 함수 사용법)
데이터 그룹화 (groupby 함수 사용법)
데이터 변환 함수 (apply, map, applymap 등)
데이터 피벗 (pivot_table 함수 사용법)
데이터 병합 (merge 함수 사용법)
데이터 분할 (split 함수 사용법)
데이터 샘플링 (sample 함수 사용법)
데이터 집계 (agg 함수 사용법)
데이터 시각화 (matplotlib, seaborn 등)
위의 전처리 과정에서 사용되는 함수와 기법에 대해서는 추후에 다른 게시글에서 자세히 다루도록 하겠습니다.

```
````

데이터 사이언티스트들이 생각하는 7단계 정석 순서

🚢 데이터 전처리 7단계 로직 (생각의 순서)
1단계: 데이터 파악 (Inspection) - "이 데이터는 무슨 사연일까?"
할 일: 전체 행/열 개수 확인, 데이터 타입(int, float, object) 확인.
생각할 점: 어떤 컬럼이 생존(y)에 가장 큰 영향을 줄까? (성별? 좌석 등급?)
주의: 텍스트 데이터(name, ticket)는 바로 쓸 수 없으니 어떻게 가공할지 미리 고민합니다.

2단계: 결측치 처리 (Handling Missing Values) - "빈칸을 무엇으로 채울까?"
할 일: isnull().sum()으로 빈칸 확인 후 채우기.
생각할 점: * 나이(Age): 단순히 전체 평균으로 채울까, 아니면 칭호(Title)별 평균으로 채울까? (나중에 한 게 훨씬 정확하죠!)
항구(Embarked): 가장 많이 탄 항구로 채울까?
주의: 결측치를 채우지 않으면 모델이 아예 돌아가지 않습니다.

3단계: 피처 엔지니어링 (Feature Engineering) - "숨은 보석 찾기"
할 일: 기존 변수를 조합해 새로운 변수 만들기. (예: Age → Age_Group, Name → Title, GroupSurvival 등)
생각할 점: 모델이 이해하기 쉽게 데이터를 가공해 주는 단계입니다. 여기서 실력 차이가 극명하게 갈립니다.

4단계: 이상치 처리 (Outlier Detection) - "튀는 놈들을 어떻게 할까?"
할 일: 말도 안 되게 높은 요금(Fare)이나 잘못 입력된 데이터 확인.
생각할 점: 이 데이터가 진짜 정보인가, 아니면 입력 오류인가?
주의: 랜덤 포레스트는 이상치에 강하지만, 그래디언트 부스팅은 민감할 수 있습니다.

5단계: 문자열 인코딩 (Encoding) - "컴퓨터 언어로 번역하기"
할 일: pd.get_dummies 등을 사용하여 글자를 숫자(0, 1)로 변환.
생각할 점: * 성별처럼 두 가지면 0과 1로.
항구나 칭호처럼 여러 개면 원-핫 인코딩으로 펼치기.
주의: 인코딩을 안 하면 아까 보신 ValueError: could not convert string to float 에러가 납니다.

6단계: 데이터 스케일링 (Scaling) - "단위 통일하기"
할 일: StandardScaler나 MinMaxScaler 사용.
생각할 점: '나이(0~80)'와 '요금(0~500)'은 단위가 다릅니다.
주의: 우리가 쓴 트리 기반 모델(RF, GBM)은 스케일링이 필수는 아닙니다. 하지만 딥러닝이나 거리를 재는 모델(KNN, SVM)에서는 목숨만큼 중요합니다!

7단계: 검증 준비 (Data Splitting) - "공정하게 시험 볼 준비"
할 일: train_test_split 또는 KFold 설정.
생각할 점: 아까 배우신 것처럼 데이터를 잘 섞었는가(shuffle=True)?
주의: 학습 데이터로 시험까지 보면 점수가 조작되니 반드시 나누어야 합니다.

💡 요약: "전처리의 핵심 철학"
전처리를 할 때 항상 머릿속에 두어야 할 질문은 딱 하나입니다.
"내가 만든 이 변수가 모델이 정답(생존 여부)을 맞히는 데 진짜 힌트가 되는가?"
단순히 Age를 넣는 것보다 Age_Group으로 묶어주는 것이 모델에겐 훨씬 친절한 설명이 됩니다.

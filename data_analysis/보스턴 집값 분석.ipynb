### 🏘️ 보스턴 주택 가격 데이터셋 (Boston Housing Dataset) 변수 설명
| 변수명 | 설명 - 상세 내용 |
|:---|:-------------|
| **CRIM** | 범죄율 - 자치시(town)별 1인당 범죄 발생률 |
| **ZN** | 주거지 비율 - 25,000 평방피트를 초과하는 거주 지역의 비율 |
| **INDUS** | 비소매 상업지 비율 - 구역별 비소매 상업 지역이 차지하는 면적 비율 |
| **CHAS** | 찰스강 더미 변수 - 강변에 위치하면 1, 아니면 0 (강 경계에 인접해 있는지 여부) |
| **NOX** | 일산화질소 농도 - 공기 중 일산화질소 농도 (1,000만 분의 1 단위) |
| **RM** | 방 개수 - 주택당 평균 방의 개수     |
| **AGE** | 노후 주택 비율 - 1940년 이전에 건축된 소유 주택의 비율 |
| **DIS** | 고용센터 거리 - 보스턴 5대 고용 센터까지의 가중 거리 |
| **RAD** | 고속도로 접근성 - 방사형 고속도로까지의 접근성 지수 |
| **TAX** | 재산세율 - $10,000당 재산세율 |
| **PTRATIO** | 학생,교사 비율 - 자치시별 학생 대비 교사 수 비율 |
| **B** | 흑인 인구 비율 - $1000(Bk - 0.63)^2$ (Bk는 자치시별 흑인 비율) |
| **LSTAT** | 하위 계층 비율 - 인구 중 하위 계층의 비율 (%) |
| **MEDV** | 주택 가격 (Target) - 본인 소유 주택 가격의 중앙값 (단위: $1,000) |

- CRIM: 자치시별 1인당 범죄 발생률
- ZN: 25,000 평방피트 초과 주거지 비율
- INDUS: 비소매 상업 지역 면적 비율
- CHAS: 찰스강 인접 여부 (1/0)
- NOX: 일산화질소 농도
- RM: 주택당 평균 방 개수
- AGE: 1940년 이전 건축된 노후 주택 비율
- DIS: 5대 고용 센터까지의 가중 거리
- RAD: 방사형 고속도로 접근성 지수
- TAX: $10,000당 재산세율
- PTRATIO: 자치시별 학생 대비 교사 수 비율
- B: 흑인 인구 비율 관련 지수
- LSTAT: 인구 중 하위 계층 비율 (%)
- MEDV: [Target] 주택 가격 중앙값

```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt 
import koreanize_matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('boston.csv')
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
# 방의 개수가 많으면 가격도 비싸다 
plt.plot(df.iloc[:,5].values,df.iloc[:,-1].values)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(df.iloc[:,-2].values,df.iloc[:,-1].values, alpha=0.4, s=10, color='#4C72B0')

plt.title('보스턴 하위 계층 비율과 주택 가격의 관계', fontsize=15, pad=15)
plt.xlabel('하위 계층 비율', fontsize=12)
plt.ylabel('주택 가격 (단위: 1000달러)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 전체 컬럼 목록 확인
cols = df.columns.tolist()

# 0번부터 4번 컬럼까지 5개만 먼저 보기
sns.pairplot(df[cols[0:5]])
plt.show()

# 그다음 5개 보기
sns.pairplot(df[cols[5:10]])
plt.show()
```

```python
sns.pairplot(df, 
             vars=cols[0:5],   # 그릴 컬럼들 (여기 'MEDV' 안 들어가도 됨)
             hue='MEDV',      # 색깔 기준 (데이터프레임 안에만 있으면 됨)
             palette='viridis') # 색깔 테마 (옵션)
plt.show()
sns.pairplot(df, 
             vars=cols[5:10],   # 그릴 컬럼들 (여기 'MEDV' 안 들어가도 됨)
             hue='MEDV',      # 색깔 기준 (데이터프레임 안에만 있으면 됨)
             palette='viridis') # 색깔 테마 (옵션)
plt.show()
```

```python
# 1. 색상 테마(Palette) 내 마음대로 바꾸기
# 기본 색상이 질린다면 palette 옵션으로 분위기를 확 바꿀 수 있습니다.

# 'magma', 'viridis', 'rocket', 'deep' 등 취향껏!
sns.pairplot(df, vars=cols[0:5], hue='MEDV', palette='husl') 
plt.show()
```

```python
# 2. 점이 너무 겹친다면? 투명도(alpha) 조절
# 데이터 양이 많으면 색깔이 겹쳐서 구분이 안 될 때가 있어요. 
# 그럴 땐 투명도를 주면 겹치는 부분이 진하게 보여서 분포가 더 잘 보입니다.

# plot_kws 안에 alpha 값을 0~1 사이로 줍니다.
sns.pairplot(df, vars=cols[0:5], hue='MEDV', 
             plot_kws={'alpha': 0.5}) 
plt.show()
```

```python
# 3. 대각선 그래프 스타일 바꾸기 (diag_kind)
# 가운데 대각선에 나오는 히스토그램이 심심하다면, 부드러운 곡선(KDE)으로 바꿀 수 있어요. 
# 색깔별로 밀도가 그려져서 훨씬 고급스러워 보입니다.
sns.pairplot(df, vars=cols[0:5], hue='MEDV', diag_kind='kde')
plt.show()
```

```python
# 1. 추천하는 '한 우물' 팔레트
# 가장 깔끔하게 "낮음 -> 높음"이 밝기 차이로 보이는 테마들이에요. 이름 뒤에 **_r**을 붙이면 진하기 순서가 반전됩니다.

# 테마 이름,색상 느낌,추천 코드
# Blues,연한 하늘색 → 진한 남색,palette='Blues'
# Gresy,연한 회색 → 검정색 (완전 깔끔),palette='Greys'
# YlOrBr,노랑 → 주황 → 갈색 (가을 느낌),palette='YlOrBr'
# mako,Seaborn 전용 (세련된 어두운 계열),palette='mako'

# 1. 가격을 'Low', 'Medium', 'High' 3단계로 나누는 임시 컬럼 생성
df['price_grade'] = pd.qcut(df['MEDV'], q=3, labels=['Low', 'Medium', 'High'])

# 2. 'price_grade'를 hue로 사용 (연속적일 때보다 훨씬 명확함!)
# 'Blues'를 쓰면 연한 파랑에서 진한 파랑으로 단계가 딱 나뉩니다.
sns.pairplot(df, vars=cols[0:5], hue='price_grade', palette='Blues')  # 여기가 핵심! 
plt.show()
```

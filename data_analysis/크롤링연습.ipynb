```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. 데이터를 담을 리스트 준비
all_quotes = []

# 2. 1페이지부터 10페이지까지 반복 (마지막 페이지 확인 필요)
for page in range(1, 11):
    url = f'http://quotes.toscrape.com/page/{page}/'
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')
        quotes = soup.select('.quote')
        
        # 페이지 내의 각 명언 데이터 추출
        for q in quotes:
            text = q.select_one('.text').text
            author = q.select_one('.author').text
            # 태그 정보도 가져오고 싶다면?
            tags = [tag.text for tag in q.select('.tag')]
            
            # 딕셔너리 형태로 리스트에 추가
            all_quotes.append({
                'Quote': text,
                'Author': author,
                'Tags': ', '.join(tags)  # 리스트를 문자열로 변환
            })
        print(f"{page}페이지 완료 (누적 {len(all_quotes)}개)")
    else:
        print(f"{page}페이지 호출 실패: {response.status_code}")
        break # 오류 발생 시 중단

# 3. 데이터프레임 생성
df = pd.DataFrame(all_quotes)

# 4. 결과 확인 및 저장
print("\n--- 최종 데이터프레임 상단 5행 ---")
print(df.head())

# CSV 파일로 저장 (한글이 없지만 인코딩 습관을 위해 utf-8-sig 권장)
df.to_csv('quotes_all.csv', index=False, encoding='utf-8-sig')
print("\n'quotes_all.csv' 파일로 저장되었습니다.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. 데이터를 담을 리스트와 시작 페이지 설정
all_quotes = []
page = 1

print("크롤링을 시작합니다...\n")

while True:
    # URL 설정
    url = f'http://quotes.toscrape.com/page/{page}/'
    response = requests.get(url)
    
    # 연결 확인
    if response.status_code != 200:
        print(f"{page}페이지를 불러오지 못했습니다.")
        break
        
    soup = BeautifulSoup(response.text, 'lxml')
    quotes = soup.select('.quote')
    
    # 데이터 추출 (tags 제외)
    for q in quotes:
        text = q.select_one('.text').text
        author = q.select_one('.author').text
        
        all_quotes.append({
            'Quote': text,
            'Author': author
        })
        
    print(f"현재 {page}페이지 완료 (수집된 명언: {len(all_quotes)}개)")
    
    # 다음 페이지가 있는지 확인 (Next 버튼 여부 조사)
    next_button = soup.select_one('li.next') # li 태그의 next 클래스
    
    if next_button:
        page += 1  # 다음 페이지로 번호 증가
    else:
        print("\n마지막 페이지에 도달했습니다. 종료합니다.")
        break

# 2. 데이터프레임 생성 및 확인
df = pd.DataFrame(all_quotes)

# 주피터 노트북에서 표 형태로 바로 확인
display(df.head())

# CSV 저장 (인덱스 제외)
df.to_csv('quotes_simple.csv', index=False, encoding='utf-8-sig')
```

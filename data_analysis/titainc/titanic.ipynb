##### 결측치 처리
- 나이는 평균값으로 채우고 확인

```python
mean_age = df['Age'].mean()
mean_age
```

```python
df['Age'] = df['Age'].fillna(mean_age)
df.isnull().sum()
```

#### 성별 그룹화하여 생존여부 확인
- 성별을 그룹화하고 생존의 평균 구하여 확인

```python
survived_by_sex = df.groupby('Sex')['Survived'].mean()
survived_by_sex
```

```python
import requests
from bs4 import BeautifulSoup
import lxml
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
url = "https://comic.naver.com/webtoon?tab=wed"
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.get(url)
time.sleep(2) 


articles = driver.find_elements(By.CSS_SELECTOR, 'div.ContentList__info_area--bXx7h')
    
time.sleep(2)

# 5. 데이터 추출 및 출력
for i, article in enumerate(articles, 1):
    title = article.find_elements(By.CSS_SELECTOR, "span.text")                  # 화면에 보이는 텍스트(제목) 가져오기
    author = article.find_element(By.CSS_SELECTOR, "a.ContentAuthor__author--CTAAP").text  # 숨겨진 속성(링크 주소) 가져오기

    print(f"{i}.  웹툰 제목 : {title[0].text}")
    print(f"    저자 : {author}")
    print(f"    별점 : {title[1].text}  \n {'='*50} ")

driver.quit()
print("\n수집이 완료되었습니다!")
```

```python
import requests
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import os
from dotenv import load_dotenv

# .env 파일 로드
load_dotenv()
# 1. API 요청 URL 및 파라미터 설정
url = 'http://apis.data.go.kr/1613000/RTMSDataSvcAptTradeDev/getRTMSDataSvcAptTradeDev'
# 본인의 발급받은 Decoding 키를 입력하세요.
service_key = os.getenv('DATA_GO_KR_API_KEY')

params = {
    'serviceKey':service_key,
    'pageNo':'1',
    'numOfRows':'100',
    'LAWD_CD':'11110',
    'DEAL_YMD':'202312'
}

# 2. API 요청 및 응답 확인
response = requests.get(url, params=params)


# 3. XML 데이터를 데이터프레임으로 변환

```

```python
import pandas as pd
import xml.etree.ElementTree as ET

# 1. response.content 또는 제공된 문자열 데이터를 사용합니다.
# response = requests.get(url, params=params)
# xml_data = response.content

# 여기서는 예시를 위해 직접 데이터를 변환합니다.
xml_data = response.content  

# 2. XML 파싱
root = ET.fromstring(xml_data)

# 3. <item> 태그 내의 데이터를 추출하여 리스트에 담기
items_list = []
for item in root.findall('.//item'):
    item_dict = {}
    for child in item:
        item_dict[child.tag] = child.text
    items_list.append(item_dict)

# 4. 데이터프레임 생성
df = pd.DataFrame(items_list)

# 데이터 확인
print(f"총 {len(df)}건의 데이터가 변환되었습니다.")
display(df.head())
```

```python
df.to_csv('APT_Feature.csv', encoding='utf-8-sig')
```

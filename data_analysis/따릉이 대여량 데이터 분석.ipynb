```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import koreanize_matplotlib
## 맷플롯립 한글 설정( koreanize-matplotlib 설치해서 대체)
# matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# matplotlib.rcParams['font.size'] = 15
# matplotlib.rcParams['axes.unicode_minus'] = False
```

- 2026년 2월 1일부터 7일까지 1시간 간격으로 시간 데이터 생성

```python
dates = pd.date_range('2026-2-1', '2026-2-7 23:00:00' , freq='h')
dates
```

- 10부터 50미만의 dates 개수만큼의 가상 데이터 생성( 8시와 18시 출퇴근시간에 대여량이 급증 )

```python
dates.hour
```

```python
rng = np.random.default_rng(0)
counts = rng.integers(10, 50, len(dates))+(dates.hour==8)*70+(dates.hour==18)*85
counts
# 출퇴근 시간에 가중치 적용
# pd.Series(np.random.randint(0, 50, size=168))
```

- 데이터프레임으로 합치기

```python
df = pd.DataFrame({'datetime':dates, 'count':counts})
df.head()
```

#### 시간데이터 뽑아내기 - 0시부터 23시까지 시간별로 대여량의 평균 구하기

- 데이터프레임 datetime열에는 연도, 월, 일, 시간이 모두 섞여 있음. 우리가 궁금한 것은 몇시에 가장 많이 대여하는지 여부
- 시간 정보만 따로 분리해 보자 ( .dt.hour : 시간, .dt.dayofweek : 요일 )

```python
df['hour']=df['datetime'].dt.hour
df.head(30)
```

- 시간으로 그룹으로 묶고, 대여량의 평균구하기

```python
hour_avg = df.groupby('hour')['count'].mean()
hour_avg
```

#### 시간 흐름에 따른 대여량 변화를 꺾은선 그래프로 시각화

```python
import matplotlib.pyplot as plt
# 1. 그래프 크기 설정 및 스타일
plt.figure(figsize=(12, 6))

# 2. 꺾은선 그래프 그리기 (marker='o'를 넣으면 시간대별 점이 찍혀서 더 잘 보여요)
plt.plot(hour_avg.index, hour_avg.values, marker='o', linestyle='-', color='b', linewidth=2)

# 3. 그래프 제목 및 축 이름 설정
plt.title('Hourly Average Rental Count', fontsize=15)
plt.xlabel('Hour (0-23)', fontsize=12)
plt.ylabel('Average Rental Count', fontsize=12)

# 4. X축 간격 설정 (0~23시까지 모두 표시)
plt.xticks(range(0, 24))

# 5. 격자(Grid) 추가 및 출력
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
import requests
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET
import os
from dotenv import load_dotenv
```

```python
# .env 파일 로드
load_dotenv()
# 1. API 요청 URL 및 파라미터 설정
url = 'http://apis.data.go.kr/1613000/RTMSDataSvcAptTradeDev/getRTMSDataSvcAptTradeDev'
# 본인의 발급받은 Decoding 키를 입력하세요.
service_key = os.getenv('DATA_GO_KR_API_KEY')

params = {
    'serviceKey':service_key,
    'pageNo':'1',
    'numOfRows':'100',
    'LAWD_CD':'11110',
    'DEAL_YMD':'202312'
}

# 2. API 요청 및 응답 확인
response = requests.get(url, params=params)
root = ET.fromstring(response.content)

# 3. XML 데이터를 데이터프레임으로 변환
data_list = []
rng = np.random.default_rng(seed=42)
# 선택할 지역 리스트 정의
locations = ['서울', '광주', '대전', '대구', '제주']

for item in root.findall('.//item'):
    data = {
        'Transaction_Date': item.find('dealYear').text + item.find('dealMonth').text + item.find('dealDay').text,
        'dealAmount': item.find('dealAmount').text if item.find('dealAmount') is not None else None,
        'excluUseAr': item.find('excluUseAr').text if item.find('excluUseAr') is not None else None,
        'floor': item.find('floor').text if item.find('floor') is not None else None,
        'buildYear': item.find('buildYear').text if item.find('buildYear') is not None else None,
        # rng.choice를 사용하여 리스트 중 하나를 무작위 선택
        'Location': rng.choice(locations) 
    }
    data_list.append(data)

df_api = pd.DataFrame(data_list)
df_api.to_csv('df_api.csv', encoding='utf-8-sig')
df_api
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

df_api = pd.read_csv('df_api.csv').drop(columns=['Unnamed: 0'])

print("=== 원본 데이터 ===\n", df_api, "\n")

df_api['floor'] = pd.to_numeric(df_api['floor'])
df_api['buildYear'] = pd.to_numeric(df_api['buildYear'])
df_api['dealAmount'] = df_api['dealAmount'].str.replace(',','').astype(int)



## 1. 결측치 처리 (Imputation)
# Floor(층)는 중앙값으로, Built_Year(건축년도)는 최빈값으로 채웁니다.
df_api['floor'] = df_api['floor'].fillna(df_api['floor'].median())
# 최빈값(mode)은 Series를 반환하므로 [0]으로 첫 번째 값을 추출합니다.
df_api['buildYear'] = df_api['buildYear'].fillna(df_api['buildYear'].mode()[0])
print("=== 결측치 처리후 ===\n", df_api, "\n")

## 2. 범주형 변수 인코딩 (Categorical Encoding)
# Location(지역)은 순서가 없는 명목형 데이터이므로 원-핫 인코딩(One-Hot Encoding)을 적용합니다.
df = pd.get_dummies(df_api, columns=['Location'], drop_first=True, dtype=int)
print("=== 범주형 변수 인코딩 처리후 ===\n", df, "\n")

## 3. 수치형 변수 스케일링 (Feature Scaling)
# Area(면적) 변수를 모델이 잘 학습하도록 평균 0, 표준편차 1인 표준화(Standardization)를 진행합니다.
scaler = StandardScaler()
df['Area_Scaled'] = scaler.fit_transform(df[['excluUseAr']])
print("=== 수치형 변수 스케일링 처리후 ===\n", df, "\n")

## 4. 데이터 변환 및 구간화 (Transformation & Binning)
# 거래금액(Price)은 단위가 크고 우측으로 꼬리가 긴 분포를 가지기 쉬우므로 로그 변환을 합니다.
df['Price_Log'] = np.log1p(df_api['dealAmount'])

# 건축년도(Built_Year)를 연대별로 묶어 범주형 변수로 만듭니다.
bins = [1900, 1999, 2009, 2029]
labels = ['1990s_and_befor', '2000s', '2010s_and_after']
df['Built_Decade'] = pd.cut(df['buildYear'], bins=bins, labels=labels)
print("=== 수치형 변수 스케일링 처리 후 ===\n", df, "\n")

## 5. 파생 변수 생성 (Feature Creation)
# Transaction_Date에서 연/월 파생 변수를 추출합니다.
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
df['Trans_Year'] = df['Transaction_Date'].dt.year
df['Trans_Month'] = df['Transaction_Date'].dt.month

# 현재 시점을 기준으로 건물의 나이(Age_of_Building)라는 새로운 핵심 피처를 만듭니다.
df['Age_of_Building'] = 2026 - df['buildYear']
print("=== 파생 변수 생성 처리 후 ===\n", df, "\n")
df_api.head()
```

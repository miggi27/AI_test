```python
import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd
```

```python
all_quotes = []
page = 1
while True:
    url = f'http://quotes.toscrape.com/page/{page}/'
    response = requests.get(url)
    if response.status_code != 200:
        break
    soup = BeautifulSoup(response.text, 'lxml')
    quotes = soup.select('.quote')
    for q in quotes:
        text = q.select_one('.text').text
        author = q.select_one('.author').text
        all_quotes.append({
            'Quote': text,
            'Author': author
        })
    next_button = soup.select_one('li.next')
    if next_button:
        page += 1
    else:
        break
df = pd.DataFrame(all_quotes)
df.head()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
all_quotes = []
page = 1
try:
    while True:
        url = f'https://comic.naver.com/bestChallenge?page={page}'
        driver.get(url)
        time.sleep(1.5)
        quotes = driver.find_elements(By.CLASS_NAME, 'ChallengeListItem__info_area--hnJTz')
        if not quotes:
            break
        for q in quotes:
            try:
                title=q.find_element(By.CLASS_NAME, 'ContentTitle__title--e3qXt').text
                author=q.find_element(By.CLASS_NAME, 'ContentAuthor__author--CTAAP').text
                rating=q.find_element(By.CLASS_NAME, 'Rating__star_area--dFzsb').text.replace('별점', '').strip()
                views=q.find_element(By.CLASS_NAME, 'Rating__view_area--GQb_S').text.replace('조회수', '').strip()

                all_quotes.append({
                    'Title': title,
                    'Author': author,
                    'Rating': rating,
                    'Views': views,
                })
            except:
                continue
        next_btn = driver.find_elements(By.CLASS_NAME, 'PageNavigation__button_next--B9N_1')
        if next_btn:
            is_disabled = next_btn[0].get_attribute('disabled')
            if is_disabled is not None:
                break
        page += 1
except Exception as e:
    print(f"에러 발생: {e}")
finally:
    driver.quit()
    print(f"총 {len(all_quotes)}개의 데이터를 수집했습니다.")
df = pd.DataFrame(all_quotes)
df.to_csv('Webtoon_All.csv', index=False, encoding='utf-8-sig')
df           
```

```python
import requests
from bs4 import BeautifulSoup
import lxml
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
# url = "https://comic.naver.com/webtoon?tab=wed" # 수요일
url = "https://comic.naver.com/webtoon?tab=thu" # 목요일
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless') # 창 없이 실행하고 싶을 때 주석 해제
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.get(url)
# time.sleep(2) 한 페이지에서 한번 가져오는 것이기에 필요 없어보임
articles = driver.find_elements(By.CSS_SELECTOR, 'div.ContentList__info_area--bXx7h')
# time.sleep(2)
for i, article in enumerate(articles, 1):
    # 화면에 보이는 텍스트(제목) 가져오기 
    title = article.find_elements(By.CSS_SELECTOR, "span.text")
    # 숨겨진 속성(링크 주소) 가져오기
    author = article.find_element(By.CSS_SELECTOR, "a.ContentAuthor__author--CTAAP").text
    print(f"{i}. 웹툰 제목 : {title[0].text}")
    print(f" 저자 : {author}")
    print(f" 별점 : {title[1].text} \n {'='*80}")
driver.quit()
print("\n수집이 완료되었습니다.")
```

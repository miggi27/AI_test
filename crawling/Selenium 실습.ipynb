```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

print("브라우저를 실행합니다...")

# 1. 자동으로 크롬 드라이버를 설치하고 브라우저 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 2. 네이버 뉴스 검색 페이지로 이동 (검색어: 파이썬)
keyword = "파이썬"
url = f"https://search.naver.com/search.naver?where=news&query={keyword}"
driver.get(url)

# 3. 브라우저가 뉴스를 모두 띄울 때까지 2초간 대기 (매우 중요!)
time.sleep(2) 

print(f"'{keyword}' 뉴스 수집을 시작합니다.\n" + "="*40)

# 4. 뉴스 제목 요소 찾기 (BeautifulSoup의 select와 유사함)
articles = driver.find_elements(By.CSS_SELECTOR, 'span.sds-comps-text-type-headline1')

# 5. 데이터 추출 및 출력
for i, article in enumerate(articles, 1):
    title = article.text                  # 화면에 보이는 텍스트(제목) 가져오기
    link = article.get_attribute('href')  # 숨겨진 속성(링크 주소) 가져오기
    
    print(f"{i}. {title}")
    print(f"   🔗 {link}")

# 6. 작업이 끝나면 브라우저 닫기
driver.quit()
print("="*40 + "\n수집이 완료되었습니다!")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

print("브라우저를 실행합니다...")

# 1. 자동으로 크롬 드라이버를 설치하고 브라우저 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 2. 네이버 뉴스 검색 페이지로 이동 (검색어: 파이썬)
keyword = "파이썬"
url = f"https://search.naver.com/search.naver?where=news&query={keyword}"
driver.get(url)

# 3. 브라우저가 뉴스를 모두 띄울 때까지 2초간 대기
time.sleep(2) 

print(f"'{keyword}' 뉴스 수집을 시작합니다.\n" + "="*40)

# 4. 뉴스 제목 요소 찾기

last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    articles = driver.find_elements(By.CSS_SELECTOR, 'a.djJH4NJWd85iOh4w2PWK')
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        break

    last_height = new_height

    # 5. 데이터 추출 및 출력
    for i, article in enumerate(articles, 1):
        title = article.text                  # 화면에 보이는 텍스트(제목) 가져오기
        link = article.get_attribute('href')  # 숨겨진 속성(링크 주소) 가져오기
        
        print(f"{i}. {title}")
        print(f"   🔗 {link}")

# 6. 작업이 끝나면 브라우저 닫기
driver.quit()
print("="*40 + "\n수집이 완료되었습니다!")
```
